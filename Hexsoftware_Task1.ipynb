```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset from seaborn
iris = sns.load_dataset('iris')

# Display the first few rows
print(iris.head())
iris.info()
```

```python
# Check for missing values
print(iris.isnull().sum())

# Simulating missing values in 'sepal_length'
iris.loc[iris.sample(frac=0.1).index, 'sepal_length'] = None

# Check for duplicates
print(f"Duplicate rows: {iris.duplicated().sum()}")

# Remove duplicates if any (not necessary for Iris, but for practice)
iris.drop_duplicates(inplace=True)

# Check for data types
print(iris.dtypes)
```

```python

# Summary statistics for numeric columns only
print(iris.describe())

# Group by species to explore species-specific statistics for numeric columns
print(iris.groupby('species').mean())

# Check for skewness, excluding the 'species' column
print(iris.drop(columns='species').skew())
```

```python
#plotting the histogram

iris.hist(bins=15, figsize=(10, 6),color='red')
plt.suptitle('Feature Distributions of Iris Dataset')
plt.show()
```

```python
# Scatter plot of sepal length vs sepal width, colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='deep', s=100)

# Add titles and labels
plt.title('Sepal Length vs Sepal Width (by Species)', fontsize=14)
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(iris.drop(columns='species').corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Features')
plt.show()

```

```python
# Boxplot for Sepal Length grouped by species
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris, x='species', y='sepal_length',color='green')
plt.title('Boxplot of Sepal Length by Species', fontsize=14)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal Length (cm)', fontsize=12)
plt.show()
```

```python
# Pairplot to visualize pairwise relationships
sns.pairplot(iris, hue='species', palette='bright')
plt.suptitle('Pairplot of Iris Features', y=1.02)
plt.show()
```
