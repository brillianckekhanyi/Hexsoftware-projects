```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

### datasets and predictions for demonstration

```python
y_test_strong_binary = [0]*50 + [1]*50
y_pred_strong_binary = [0]*45 + [1]*5 + [0]*3 + [1]*47

y_test_imbalance_binary = [0]*100 + [1]*100
y_pred_imbalance_binary = [0]*80 + [1]*20 + [0]*30 + [1]*70

y_test_multiclass = [0]*45 + [1]*40 + [2]*50
y_pred_multiclass = [0]*38 + [1]*5 + [2]*2 + [0]*3 + [1]*30 + [2]*7 + [0]*2 + [1]*10 + [2]*38


accuracy = accuracy_score(y_test_strong_binary, y_pred_strong_binary)
precision = precision_score(y_test_strong_binary, y_pred_strong_binary, average='weighted')
recall = recall_score(y_test_strong_binary, y_pred_strong_binary, average='weighted')
f1 = f1_score(y_test_strong_binary, y_pred_strong_binary, average='weighted')
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

## Function to plot confusion matrix with heatmap visualization


```python
def plot_confusion_matrix(y_test, y_pred, title):
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{title} - Confusion Matrix")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Plotting confusion matrices for each example scenario
plot_confusion_matrix(y_test_strong_binary, y_pred_strong_binary, "Binary Classification - Strong Classifier")
plot_confusion_matrix(y_test_imbalance_binary, y_pred_imbalance_binary, "Binary Classification - Class Imbalance Issue")
plot_confusion_matrix(y_test_multiclass, y_pred_multiclass, "Multiclass Classification - Some Misclassifications")
```

##  Classification report


```python
print("Classification Report:")
print(classification_report(y_test_strong_binary, y_pred_strong_binary))
```
